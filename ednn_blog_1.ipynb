# Extensive deep neural network tutorial

Here we will build an Extensive Deep Neural Network (EDNN) capable of predicting extensive physical properties on two-dimensional images.  We will demonstrate how to use the trained model to achieve $\mathcal{O}(N)$ scaling with respect to the image size.

This technique was first introduced in "Extensive Deep Neural Networks" by I. Luchak, K. Mills, K. Ryczko, et. al.
(https://arxiv.org/abs/1708.06686)



## 1) Count the jellybeans

An obvious extensive property is the number of visible objects within an image.  We will tackle an over-simplified model of the common carnival game where the player must guess the number of jellybeans in a container by visually inspecting the container. We will reduce the complexity of the problem significantly by learning to count the number of 1s in a field of zeros; an obviously easy task for both humans and a computer.  Clearly we don't need a neural network for this task, let alone an extensive neural network, but it will demonstrate well the concepts we wish to extend to larger, more complex problems.

```python
#First, some imports
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage.filters import gaussian_filter
import progressbar
%matplotlib inline
```

We first must generate some data.  We will do this by generating a bunch of $64\times 64$ images of zeros, then randomly flip between 10 and 200 indices per image. We'll then return the training data (the image) and the labels (the number of bright pixels).  We'll then plot a few of the images to make sure they look nice.

```python
def make_data(N, L):
    image = np.zeros((N, L, L))
    labels = np.random.randint(10,200, (N))
    
    for i in range(N):
        for num_points in range(labels[i]):
            x = np.random.randint(0,L)
            y = np.random.randint(0,L)
            image[i,x,y] = 1.0
        image[i] = gaussian_filter(image[i],0.8)
    return image, labels.reshape((-1,1))


data, labels =  make_data(25,64)

fig, axs = plt.subplots(5,5,figsize=(10,10))

for i in range(5):
    for j in range(5):
        axs[i][j].imshow(data[5*i+j], cmap='Spectral')
        axs[i][j].set_title('{0}'.format(labels[5*i+j][0]))
        axs[i][j].axis("off")
```

Next we need to decide on the focus and context that we will use.  For this case, where there is no inter-pixel interactions, this choice doesn't matter at all, but we'll demonstrate it nontheless.

Since the focus regions are non-overlapping, our image size must be evenly divisible by our choice of focus region,  i.e. $L\%f = 0$. Since our image size is $L=64$, we'll use a focus size of $f=16$.

The size of our context defines the length-scale of interactions, and since there are no interactions, our choice of $c$ does not matter for this case.  We'll choose $c=4$.

Therefore the neural network that we will ultimately build will receive a tile of size $f + 2c = 24$.

```python
L = 64
f = 16
c = 4
```

Next we'll define a few functions to help us.  `roll` is merely a 2-dimensional TensorFlow implementation of `np.roll()`, and `slice()` just takes a starting coordinate pair and width/height and slices a 2d tensor appropriately.  The real magic is in the `ednn_split()` function, which takes an input image of size `[L, L]`, and busts it up into the appropriate number of tiles, given the globally defined focus and context. 

```python
def roll(in_, num, axis):
    """author: Kyle Mills"""
    D = tf.transpose(in_, perm=[axis, 1-axis])  #if axis=1, transpose first
    D = tf.concat([D[num:, :], D[0:num, :]], axis=0)
    return tf.transpose(D, perm=[axis, 1-axis]) #if axis=1, transpose back

def slice_(in_, x1, y1, w, h): 
    """author: Kyle Mills"""
    return in_[x1:x1+w, y1:y1+h]

def ednn_split(in_): 
    """author: Kyle Mills"""
    l = L
    #alternatively, a (probably better) way to get l is to do:
    #   l = in_.get_shape().as_list()[0]
   
    tiles = []
    for iTile in range(l/f):
        for jTile in range(l/f):
            #calculate the indices of the centre of this tile (i.e. the centre of the focus region)
            cot = (iTile*f + f/2, jTile*f + f/2) #centre of tile
            foc_centered = in_ 
            #shift the picture, wrapping the image around,
            #so that the focus is centered in the middle of the image
            foc_centered = roll(foc_centered, l/2-cot[0], 0)
            foc_centered = roll(foc_centered, l/2-cot[1], 1)
            #Finally slice away the excess image that we don't want to appear in this tile
            final = slice_(foc_centered, l/2-f/2-c, l/2-f/2-c, 2*c+f, 2*c+f)
            tiles.append(final)
            
    return tf.expand_dims(tiles, axis=3)
```

Next we will define the neural network.  Notice that this is done no differently than a normal neural network.  The thing to keep in mind is that the `_in` will be a batch of tiles, with size `f+2c`.  Here I use a simple 4-layer fully-connected neural network.

```python
def NN(_in):
    tile_size = f + 2*c
    _in = tf.reshape(_in, (-1, tile_size**2))
    lay1 = tf.contrib.layers.fully_connected(_in, 64)
    lay2 = tf.contrib.layers.fully_connected(lay1, 128)
    lay3 = tf.contrib.layers.fully_connected(lay2, 128)
    lay4 = tf.contrib.layers.fully_connected(lay3, 1, activation_fn=None)
    return lay4
```

Now we'll build the graph, with some tricks that you might not be familiar with if your background is traditional deep neural networks.  First the code, then an explanation of the process:

```python
#data comes in a [ batch * L * L ] tensor, and labels a [ batch * 1] tensor
x = tf.placeholder(tf.float32, (None, L, L), name='input_image')
y = tf.placeholder(tf.float32, (None, 1))

#Then the EDNN-specific code:
tiles = tf.map_fn(ednn_split, x, back_prop=False)
tiles = tf.transpose(tiles, perm=[1,0,2,3,4])
output = tf.map_fn(NN, tiles, back_prop=True)
output = tf.transpose(output, perm=[1,0,2])
predicted = tf.reduce_sum(output, axis=1)


#define the loss function
loss = tf.reduce_mean(tf.square(y-predicted))

#create an optimizer, a training op, and an init op
optimizer = tf.train.AdamOptimizer(learning_rate=0.001)
train_step = optimizer.minimize(loss)
init = tf.global_variables_initializer()
```

```python
sess = tf.InteractiveSession()
sess.run(init)
```

We will get 1000 data points, and split it into training and testing datasets with a 9/1 ratio:

```python
data, labels = make_data(1000, 64)
train_data = data[:900]
test_data = data[900:]
train_labels = labels[:900]
test_labels = labels[900:]
```

```python
BATCH_SIZE = 100

bar = progressbar.ProgressBar()
for epoch in bar(range(100)):
    for batch in xrange(train_data.shape[0] / BATCH_SIZE):
        _, loss_val = sess.run([train_step, loss],
                               feed_dict={
                                    x: train_data[batch*BATCH_SIZE:batch*(BATCH_SIZE+1)],
                                    y: labels[batch*BATCH_SIZE:batch*(BATCH_SIZE+1)]
                                }
                              )
```

```python
prediction = sess.run(predicted, feed_dict={x: test_data})
```

```python
fig, ax = plt.subplots(1,1, figsize=(5,5))

ax.plot(test_labels.flatten(), prediction.flatten(), '.', color='#821122')
ax.plot(np.arange(10,200), np.arange(10,200), '--', color='gray')
ax.set_xlabel("True number of jellybeans")
ax.set_ylabel("Predicted number of jellybeans")
ax.grid(alpha=0.3)
```

It works!  Now obviously that was too easy of a problem; we could have used a normal neural network for that problem. However, it's good to start small, and we now know how to set up an extensive deep neural network to take a 2d image as input and predict a label.
